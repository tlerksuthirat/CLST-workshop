```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Kaladp_data = pd.read_excel("Kaladp_expression_FPKM_biorep_primary_matureLeaf (1).xlsx", index_col=0)
```

```python
Kaladp_data.head()
```

```python
# 1. Display basic info
#print(Kaladp_data.info())

# 2. Check for missing values
#print(Kaladp_data.isnull().sum())

# 3. Summary statistics
summary_stats = Kaladp_data.describe()
print(Kaladp_data.describe())
# Export to CSV
summary_stats.to_csv('kaladp_summary_stats.csv')
```

```python
# Print all column names
print(Kaladp_data.columns)
```

```python
# Calculate the mean for every three columns
means_table = pd.DataFrame()
for i in range(0, len(Kaladp_data.columns), 3):
    end = min(i + 3, len(Kaladp_data.columns))
    if end -i > 0:
        means_table['Mean_' + str(i // 3 + 1)] = Kaladp_data.iloc[:, i:end].mean(axis=1)


# Export the table to a new Excel file
means_table.to_excel("means_table.xlsx")

means_table.head()
```

```python
# prompt: how to normalize each column
from sklearn.preprocessing import MinMaxScaler

# Assuming 'means_table' DataFrame is already created as in the previous code.

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Select only numerical columns for normalization
numerical_cols = means_table.select_dtypes(include=['number']).columns

# Fit and transform the numerical columns
normalized_data = scaler.fit_transform(means_table[numerical_cols])

# Create a new DataFrame with normalized data
normalized_df = pd.DataFrame(normalized_data, columns=numerical_cols, index=means_table.index)

# Display the first few rows of the normalized DataFrame
normalized_df.head()
```

```python
from tslearn.clustering import TimeSeriesKMeans

# Create and fit the K-medoids model with DTW
model = TimeSeriesKMeans(n_clusters=5,  # Experiment with different numbers of clusters
                        metric="dtw", 
                        max_iter=10, 
                        random_state=0)  
clusters = model.fit_predict(normalized_df)
gene_clusters = pd.Series(clusters, index=normalized_df.index)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(normalized_df.iloc[model.labels_.argsort()],  # Order rows by cluster
            cmap="coolwarm", 
            xticklabels=True, 
            yticklabels=False, 
            cbar_kws={'label': 'Normalized Gene Expression'}) 

plt.title("Gene Expression Heatmap (Clustered with DTW)")
plt.xlabel("Time Points")
plt.show()
```
