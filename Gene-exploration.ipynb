```python
Kfed_TF = pd.read_csv('Kfed_TF_fromiTAK.csv', names=['Transcript', 'Gene', 'AA', 'mRNA'])
```

```python
Kfed_TF.head()
```

```python
# Count unique values for each column
for col in Kfed_TF.columns:
    unique_count = Kfed_TF[col].nunique()
    print(f"Column '{col}': {unique_count} unique values")
```

```python
Kfed_TF_sorted = Kfed_TF.sort_values(by=['Gene','AA','mRNA','Transcript'], ascending=True)
```

```python
Kfed_TF_sorted.head()
```

```python
# Remove '.p' from the 'Function' column
Kfed_TF_sorted['Transcript'] = Kfed_TF_sorted['Transcript'].astype(str).str.replace('.p', '', regex=False)

# Delete columns 'AA' and 'mRNA'
Kfed_TF_sorted = Kfed_TF_sorted.drop(['AA', 'mRNA'], axis=1)

# Rename columns
Kfed_TF_sorted = Kfed_TF_sorted.rename(columns={
    'Transcript': 'TranscriptID',
    'Gene': 'Function'
   })

# Add a new column named 'TF' with the value 'TF' in all rows
Kfed_TF_sorted['Group'] = 'TF'

Kfed_TF_sorted.to_csv('Kfed_TF_sorted.csv', index=False)
```

```python
files_to_import = ['Carbox_ID.csv', 'CAZ_ID.csv','Circadian_ID.csv','Decarbox_ID.csv','stomata_ID.csv']

# Create an empty list to store the data
data = []

# Loop through the list of files
for file in files_to_import:
  try:
    # Read the CSV file, keeping only 'TranscriptID' and 'Function' columns
    df = pd.read_csv(file, usecols=['TranscriptID', 'Function'])

    # Add a 'Group' column with the file name (without extension)
    df['Group'] = file.split('.')[0]

    # Append the data to the list
    data.append(df)
  except FileNotFoundError:
    print(f"Error: File '{file}' not found.")

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(data, ignore_index=True)

# Now you have the combined_df with 'TranscriptID', 'Function', and 'Group' columns
print(combined_df)
```

```python
# Append Kfed_TF_sorted to combined_df
appended_df = pd.concat([combined_df, Kfed_TF_sorted], ignore_index=True)
```

```python
appended_df.to_csv('annotate_transcript.csv', index=False)
```
