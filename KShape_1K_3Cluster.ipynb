```python
import numpy as np
import matplotlib.pyplot as plt
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
from tslearn.clustering import KShape

# Step 1: Scale the dataset
scaler = TimeSeriesScalerMeanVariance()
time_series_dataset_scaled = scaler.fit_transform(df_sorted)  # Ensure input shape is (n_series, n_timestamps, n_features)

# Step 2: Fit KShape with optimal k
optimal_k = 3
model = KShape(n_clusters=optimal_k, max_iter=50, random_state=0)
model.fit(time_series_dataset_scaled)

# Step 3: Get cluster labels
labels = model.labels_

# Step 4: Plot time series for each cluster
fig, axes = plt.subplots(optimal_k, 1, figsize=(10, 2 * optimal_k), sharex=True)

n_timestamps = time_series_dataset_scaled.shape[1]
time_points = np.arange(0, n_timestamps * 2, 2)  # Generate time values in 2-hour intervals

# Define the range of hours (0 - 22) for x-ticks
hours = np.arange(0, 24, 2)

for cluster_idx in range(optimal_k):
    cluster_series = time_series_dataset_scaled[np.where(labels == cluster_idx)]  # Extract series belonging to the current cluster
    ax = axes[cluster_idx]
    for series in cluster_series:
        ax.plot(time_points, series.ravel(), alpha=0.7)  # Use time_points for the x-axis
    ax.set_title(f"Cluster {cluster_idx + 1} ({len(cluster_series)} series)")
    ax.set_xlabel("Time (hours)", fontweight='bold')  # Add x-label for each subplot

    # Adjust x-ticks to match the actual number of time points
    if len(time_points) == len(hours):  # If time points and hours match
        ax.set_xticks(time_points)  # Use all time_points as ticks
        ax.set_xticklabels(hours, fontweight='bold')  # Set x-tick labels to 0 - 22 hours
    else:
        ax.set_xticks(time_points[::2])  # Show ticks at every 2nd time point
        ax.set_xticklabels(hours[:len(time_points[::2])], fontweight='bold')  # Match the length of labels

    ax.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def normalize_time_series(T):
    """Normalize a time-series by subtracting the mean and dividing by the standard deviation."""
    return (T - np.mean(T)) / np.std(T)

def ncc(T1, T2):
    """Compute the normalized cross-correlation (NCC) between two time-series."""
    return np.correlate(T1, T2, mode='valid')[0] / (np.linalg.norm(T1) * np.linalg.norm(T2))

def find_mean_shape(cluster):
    """Find the mean shape (centroid) for a cluster of time-series."""
    num_series = len(cluster)
    series_length = len(cluster[0])

    # Normalize all time-series in the cluster
    normalized_cluster = [normalize_time_series(T) for T in cluster]

    # Initialize centroid as the first time-series in the cluster
    centroid = normalized_cluster[0]

    # Iterate to update the centroid by maximizing NCC with each time-series in the cluster
    converged = False
    while not converged:
        # Calculate NCC of the centroid with all time-series in the cluster
        ncc_scores = [ncc(centroid, T) for T in normalized_cluster]

        # Find the time-series that is most similar to the current centroid
        best_match_index = np.argmax(ncc_scores)
        new_centroid = normalized_cluster[best_match_index]

        # Check for convergence: if centroid doesn't change significantly, stop iterating
        if np.allclose(new_centroid, centroid, atol=1e-6):
            converged = True
        else:
            centroid = new_centroid

    return centroid

def calculate_arithmetic_mean(cluster):
    """Calculate the arithmetic mean of a cluster of time-series."""
    return np.mean(cluster, axis=0)
```

```python
for cluster_idx in range(optimal_k):
    cluster_series = time_series_dataset_scaled[np.where(labels == cluster_idx)]  # Extract series belonging to the current cluster

    # Calculate the mean time series for the current cluster
    #mean_series = cluster_series.mean(axis=0)

    mean_shape = find_mean_shape(np.squeeze(cluster_series))
    arithmetic_mean = calculate_arithmetic_mean(cluster_series)

    # Create a new figure for each cluster
    plt.figure(figsize=(4, 2))
    for series in cluster_series:
        plt.plot(time_points, series.ravel(), alpha=0.03,color="black")  # Use time_points for the x-axis
    # Plot the mean series for the cluster
    plt.plot(time_points, mean_shape.ravel(), color='blue', linewidth=2)  # Use time_points for the x-axis
    plt.plot(time_points, arithmetic_mean.ravel(), color='red', linewidth=2)

    plt.title(f"Cluster {cluster_idx + 1} ({len(cluster_series)} series)")
    plt.xlabel("Time (hours)", fontweight='bold')  # Add x-label for each plot
    plt.xticks(time_points[::1], hours[:len(time_points[::1])], fontweight='bold')  # Set x-tick labels
    plt.xlim([0,22])
    plt.ylim([-3,3])
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
from scipy.signal import savgol_filter  # Import for smoothing

# Define smoothing parameters
window_length = 7  # Must be odd
polyorder = 2      # Polynomial order for Savitzky-Golay filter

for cluster_idx in range(optimal_k):
    cluster_series = time_series_dataset_scaled[np.where(labels == cluster_idx)]  # Extract series belonging to the current cluster

    # Calculate the mean time series for the current cluster
    mean_shape = find_mean_shape(np.squeeze(cluster_series))

    # Apply smoothing to the mean shape
    smoothed_mean_shape = savgol_filter(mean_shape.ravel(), window_length=window_length, polyorder=polyorder)

    # Create a new figure for each cluster
    plt.plot(time_points, smoothed_mean_shape, linewidth=2, label=f"Cluster {cluster_idx + 1}")  # Smoothed plot

    # Plot settings
    plt.xlabel("Time (hours)", fontweight='bold')  # Add x-label for each plot
    plt.xticks(time_points[::1], hours[:len(time_points[::1])], fontweight='bold')  # Set x-tick labels
    plt.xlim([0, 22])
    plt.ylim([-3, 3])
    plt.grid()
    plt.tight_layout()

# Show the final plot
plt.legend(loc=8,ncol=4)  # Add a legend to distinguish clusters
plt.show()
```
